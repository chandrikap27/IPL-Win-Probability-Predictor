```python
def calculate_crr(total_runs, overs_completed):
    return total_runs / overs_completed
```

```python
def calculate_rrr(target_runs, overs_remaining):
    return target_runs / overs_remaining

def calculate_balls_left(overs_remaining):
    return int(overs_remaining * 6)
```

```python
def predict_winner(team1, team2, city, team1_score, team2_score, overs_completed, wickets_left):
    runs_left = abs(team1_score - team2_score)
    balls_left = calculate_balls_left(20 - overs_completed)

    input_data=pd.DataFrame({
    'batting_team':[team1],
    'bowling_team':[team2],
    'city':[city],
    'runs_left':[runs_left],
    'balls_left':[balls_left],
    'wickets':[wickets_left],
    'total_runs_x':[team1_score],
    'crr':[calculate_crr(team1_score,overs_completed)],
    'rrr':[calculate_rrr(runs_left,20-overs_completed)]
    })

    x_test_columns = ['runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr',
          'batting_team_Chennai Super Kings', 'batting_team_Deccan Chargers',
          'batting_team_Delhi Capitals', 'batting_team_Delhi Daredevils',
          'batting_team_Kings XI Punjab', 'batting_team_Kolkata Knight Riders',
          'batting_team_Mumbai Indians', 'batting_team_Rajasthan Royals',
          'batting_team_Royal Challengers Bangalore',
          'batting_team_Sunrisers Hyderabad', 'bowling_team_Chennai Super Kings',
          'bowling_team_Deccan Chargers', 'bowling_team_Delhi Capitals',
          'bowling_team_Delhi Daredevils', 'bowling_team_Kings XI Punjab',
          'bowling_team_Kolkata Knight Riders', 'bowling_team_Mumbai Indians',
          'bowling_team_Rajasthan Royals',
          'bowling_team_Royal Challengers Bangalore',
          'bowling_team_Sunrisers Hyderabad', 'city_Abu Dhabi', 'city_Ahmedabad',
          'city_Bangalore', 'city_Bengaluru', 'city_Bloemfontein',
          'city_Cape Town', 'city_Centurion', 'city_Chandigarh', 'city_Chennai',
          'city_Cuttack', 'city_Delhi', 'city_Dharamsala', 'city_Durban',
          'city_East London', 'city_Hyderabad', 'city_Indore', 'city_Jaipur',
          'city_Johannesburg', 'city_Kimberley', 'city_Kolkata', 'city_Mohali',
          'city_Mumbai', 'city_Nagpur', 'city_Port Elizabeth', 'city_Pune',
          'city_Raipur', 'city_Ranchi', 'city_Sharjah', 'city_Visakhapatnam']

    input_encoded=pd.get_dummies(input_data, columns=categorical_columns)
    input_encoded=input_encoded.reindex(columns=x_test_columns,fill_value=0)

    probabilities = model.predict_proba(input_encoded)[0]
    team1_prob = round(probabilities[0] * 100, 2)
    team2_prob = round(probabilities[1] * 100, 2)

    if team1_prob > team2_prob:
      winner=team1
    else:
      winner=team2

    return winner, team1_prob,team2_prob
```

```python
team1 = input("Enter the name of Team 1: ")
team2 = input("Enter the name of Team 2: ")
city = input("Enter the city where they are playing: ")
team1_score = int(input("Enter the total score of Team 1: "))
team2_score = int(input("Enter the current score of Team 2: "))
overs_completed = float(input("Enter the total overs completed: "))
wickets_left = int(input("Enter the number of wickets left: "))

winner, team1_probability,team2_probability=predict_winner(team1, team2, city, team1_score, team2_score, overs_completed, wickets_left)

print("\nPredicted Winner:", winner)
print(f"Probability of {team1} winning: {team1_probability}%")
print(f"Probability of {team2} winning: {team2_probability}%")
```
